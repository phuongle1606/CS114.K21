```
f = open("data.txt", 'r')
```

```
sample = f.read()
sample = sample.split('\n')
```

```
sample
```

```
size , time = [],[]
for i in sample:
  size.append(int(i.split(',')[0]))
  time.append(float(i.split(',')[1]))
```

```
size
```

```
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(size, time, test_size = 0.2, random_state = 101)
```

```
import matplotlib.pyplot as plt
plt.scatter(size, time)
```

```
import seaborn as sns
```

```
sns.scatterplot(size, time)
```

```
import pandas as pd
x_train = pd.DataFrame (x_train)
x_test = pd.DataFrame (x_test)
```

```
x_train
```

```
from sklearn.svm import SVR
model = SVR()
model.fit(x_train, y_train)
```

```
predict = model.predict(x_test)
```

```
model.score(x_test, y_test)
```

```
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, predict)
```

```
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

```
predict=model.predict(x_test)
```

```
model.score(x_test, y_test)
```

```
cost = ((y_test - predict)**2).sum()*1/len(y_test)
```
